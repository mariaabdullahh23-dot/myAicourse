# Task 1: Exploring and Visualizing a Simple Dataset (Iris Dataset)

## Introduction
The purpose of this task is to explore and understand the Iris dataset.  
We will load the dataset, inspect its structure, check for missing values, and analyze the data using visualizations.  
This will help us see trends, distributions, and relationships between features, which is an important step before applying any machine learning model.
## Step 1: Load the Dataset
We load the Iris dataset using pandas to start our analysis.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
data = sns.load_dataset("iris")
```

```python
print(data.shape)
print(data.columns)
print(data.head())
```

## Step 2: Inspect the Dataset
Check the dataset info and statistical summary.


```python
data.info()
print(data.describe())
```

## Step 3: Visualize the Dataset
Use scatter plot, histograms, and box plot to understand relationships and detect outliers.


```python
sns.scatterplot(x="sepal_length",y="sepal_width",data=data,hue="species")
plt.show()
data.hist()
plt.show()
sns.boxplot(data=data)
plt.show()
```

## Conclusion
- The Iris dataset is clean with no missing values.  
- Scatter plots show relationships between features.  
- Histograms show feature distributions.  
- Box plots help identify outliers.  
- Overall, the dataset is ready for machine learning tasks.
